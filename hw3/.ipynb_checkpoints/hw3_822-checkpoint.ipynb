```python
import pandas
import math
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

FNAME = '../cps09mar.dta'
df = pandas.read_stata(FNAME)
```

```python
list(df.columns.values)
maritalstatus = pandas.get_dummies(df['marital']).loc[:,(1,2,3,4,5,6)]
print maritalstatus
```

```python
#part a

y = df['lwage']
X1 = df.loc[:,('educ','exp')]
X1['exp_sq'] = X1['exp']**2/100
X1 = pandas.concat([X1, maritalstatus], axis=1)
print X1
```

```python
X1 = sm.add_constant(X1)
model1 = sm.OLS(y,X1).fit()
model1.summary()
```

```python
#constrained least squares 4 = 7

#efficient minimum distance 8, 9

X2 = X1.loc[:,('educ','exp','exp_sq', 2, 3)]

X2['r1'] = X1[1].add(X1[4])
X2['r2'] = X1[5].add(X1[6])
#print X2
model2 = sm.OLS(y,X2).fit()
model2.summary()
```

```python
print 'sup'
print model2.cov_HC0
model3 = sm.GLS(y,X2,weights=model2.cov_HC0).fit()
model3.summary()
```
