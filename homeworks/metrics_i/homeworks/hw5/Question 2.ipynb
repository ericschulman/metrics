```python
import pandas
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
#Using the data in wagepan.dta estimate the wage equation using the explanatory vari-
#ables: exper, expersq, educ, union, black, poohlth. Use each of the estimators consid-
#ered in class (OLS, FE, RE) and compute appropriate standard errors. Also compute
#the Hausman test that tests the assumption that the regressors are uncorrelated with
#the \ xed e ect". Re-do the results using year dummy variables.

df = pandas.read_stata('../wagepan.dta')
print list(df)
```

```python
#set up X
X = df.loc[:,['nr','lwage','exper', 'expersq', 'educ', 'union', 'black',
              'poorhlth','d81', 'd82', 'd83', 'd84', 'd85', 'd86', 'd87']]
X = X.dropna()
X = sm.add_constant(X)
```

```python
#regular least squares
model1 = smf.OLS(X['lwage'],X[['exper', 'expersq', 'educ', 'union', 'black',
              'poorhlth']])
result1 = model1.fit()
result1.summary()
```

```python
#regular least squares with dummies
model2 = smf.OLS(X['lwage'],X[['exper', 'expersq', 'educ', 'union', 'black',
              'poorhlth','d81', 'd82', 'd83', 'd84', 'd85', 'd86', 'd87']])
result2 = model2.fit()
result2.summary()
```

```python
#fixed effects
model3 = smf.mixedlm('lwage~exper+expersq+educ+union+black+poorhlth',X,
                   groups=X['nr'])
result3 = model3.fit()
result3.summary()
```

```python
#fixed effects with dummies
model3 = smf.mixedlm('lwage~exper+expersq+educ+union+black+poorhlth+d81+d82+d83+d84+d85+d86+d87',X,
                   groups=X['nr'])
result3 = model3.fit()
result3.summary()
```
